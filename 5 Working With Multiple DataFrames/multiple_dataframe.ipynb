# `The Duplicated Index Issue`

```python
schools.loc[0]
```

pd.concat -> does not discard teh original index of the dataframes being concatenated

```python
schools.index.duplicated()
```

in pandas, we could have duplicated indices!

```python
#schools.loc[0:2]
```

`KeyError`: 'Cannot get left slice bound for non-unique label: 0'


the first fix

```python
schools.reset_index()
```

```python
schools.reset_index(drop=True, inplace = True)
```

```python
schools.head()
```

```python
schools.index.duplicated().sum()
```

the alternative approach: concat()

```python
pd.concat(dfs).drop_duplicates(subset=['School Name'])
```

the ignore_index=`True` same as `reset_index()` it just ignore the `old index` and produce new one

```python
pd.concat(dfs, ignore_index=True).drop_duplicates(subset=['School Name'])
```

```python
pd.concat(dfs, ignore_index=True).drop_duplicates(subset=['School Name']).index.duplicated().sum()
```
