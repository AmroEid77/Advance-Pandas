# `Combining Conditions`


select all the left backs -> LB


```python
players.head()
```

all left backs who are 25 or younger


```python
players[
    (players.position == 'LB') & 
    (players.age <=25) 
    ].head()
```

3 conditions : left back, 25 or younger, makret value 10+


```python
players[
    (players.position == 'LB') & 
    (players.age <=25) &
    (players.market_value.ge(10))
    ].head()
```

4 conditions : left back, 25 or younger, makret value 10+, and NOT from Arseanl and Tottenham club


```python
players[
    (players.position == 'LB') & 
    (players.age <=25) &
    (players.market_value.ge(10)) &
    ~(players.club.isin(['Arsenal', 'Tottenham']))
    ].head()
```

# `Skill Challenge`


```python
english = players.nationality =='England'
```

```python
twice_mean = players.market_value.mean() * 2
```

```python
more_than_twice_avg_market_value = players.market_value.ge(twice_mean)
```

```python
more_4000_views = players.page_views >= 4000
```

```python
new_signing =players.new_signing == 1
```
